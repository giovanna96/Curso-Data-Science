```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15, 8))
```

```python
dados = pd.read_csv('Dados/aluguel.csv', sep = ';')
dados.head(10)
```

```python
area = plt.figure() # definindo figura
# definindo duas linhas e duas colunas, subplot de 4 gráficos
g1 = area.add_subplot(2, 2, 1) # subplot gráfico 1
g2 = area.add_subplot(2, 2, 2) # subplot gráfico 2
g3 = area.add_subplot(2, 2, 3) # subplot gráfico 3
g4 = area.add_subplot(2, 2, 4) # subplot gráfico 4
```

```python
g1.scatter(dados.Valor, dados.Area)
g1.set_title('Valor X Área')

g2.hist(dados.Valor)
g2.set_title('Histograma')

dados_g3 = dados.Valor.sample(100)
dados_g3.index = range(dados_g3.shape[0])
g3.plot(dados_g3)
g3.set_title('Amostra (Valor)')

grupo = dados.groupby('Tipo')['Valor']
g4.bar(grupo.mean().index, grupo.mean().values)
g4.set_title('Valor Médio por Tipo')
```
