## Criando Novas Variáveis 

```python
import pandas as pd
```

```python
dados = pd.read_csv('Dados/aluguel_residencial.csv', sep = ';')
dados.head(10)
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados.head(10)
# inseriu uma coluna 'Valor Bruto' que é composta pela soma do 'Valor' 'Condominio' e 'IPTU' de cada Imóvel
```

```python
dados['Valor m2'] = (dados['Valor'] / dados['Area']).round(2)
dados.head(10)
# inseriu uma coluna 'Valor m2' que é composta pela divisão do 'Valor' pela 'Area' 
# função round arredonda as casas decimais
```

```python
dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
dados.head(10)
# inseriu uma coluna 'Valor Bruto m2' que é composta pela divisão do 'Valor Bruto' pela 'Area' 
```

```python
casa = ['Casa', 'Casa de Condominio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados
```

## Excluindo Variáveis

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
dados_aux.head(10)
```

```python
del dados_aux['Valor Bruto']
dados_aux.head(10)
# comando del remove variável do DataFrame
```

```python
dados_aux.pop('Valor Bruto m2')
dados_aux
# metodo pop remove variável do DataFrame
```

```python
dados.drop(['Valor Bruto', 'Valor Bruto m2'], axis = 1, inplace = True)
dados.head(10)
# função drop também remove Variável do DataFrame
```

```python
dados.to_csv('Dados/aluguel_residencial.csv', sep =';', index = False)
```
